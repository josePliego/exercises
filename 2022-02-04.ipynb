```python
import pandas as pd
import numpy as np
```

```python
dt = pd.read_csv('https://sta663-sp22.github.io/slides/data/us_rent.csv')
```

```python
dt
```

```python
dt_ratio = (dt.
 pivot(index='name', columns='variable', values='estimate')
 .assign(ratio = lambda d: d.income/d.rent)
 .reset_index()
)
```

```python
dt_ratio.sort_values('ratio', ascending = True)
```

```python
cereal = pd.read_csv('https://sta663-sp22.github.io/slides/data/cereal.csv')
cereal.head()
```

```python
cereal.groupby('type').get_group('Hot')
```

```python
cereal.groupby('mfr').agg({
    'calories': ['min', 'max'],
    'sugars': ['mean', 'median']
})
```

```python
cereal.groupby('mfr', as_index=False).agg(
    min_cal = ('calories', 'min')
)
```

```python
(cereal
 .groupby('mfr')[['calories', 'sugars', 'rating']]
 .transform(lambda x: (x - np.mean(x, axis=0))/np.std(x))
)
```
