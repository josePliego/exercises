```python
# from matplotlib import pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
sns.displot(
    data=penguins,
    x='bill_length_mm',
    hue='species',
    kind='kde',
    fill=True,
    alpha=0.5,
    aspect=1
)
```

```python
sns.pairplot(data=penguins, height=5)
```

```python
sns.color_palette()
```

```python
sns.color_palette('cubehelix', as_cmap=True)
```

```python
g = sns.PairGrid(
  penguins, hue="species",
  height=3
)
g = g.map_diag(
  sns.histplot, alpha=0.5
)
g = g.map_lower(
  sns.scatterplot
)
g = g.map_upper(
  sns.kdeplot
)
g
```
