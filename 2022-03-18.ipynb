```python
import numpy as np
import pymc3 as pm
import arviz as az
```

```python
np.random.seed(42)
n = 11
m = 6
b = 2
x = np.linspace(0, 1, n)
y = m*x + b + np.random.randn(n)
```

```python
with pm.Model() as lm:
    m = pm.Normal('m', mu=0, sd=10)
    b = pm.Normal('b', mu=0, sd=10)
    sigma = pm.HalNormal('sigma', sd=10)
    
    likelihood = pm.Normal('y', mu=m*x + b, sd=sigma, observed=y)
```
